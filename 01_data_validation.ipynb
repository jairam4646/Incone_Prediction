### Missing Value Findings

Missing values are represented by the '?' character rather than null values.

The following columns contain missing values:

- workclass
- occupation
- native_country

Occupation has the highest missingness, followed by workclass and native_country.

Missing values are present in both training and test datasets and will be handled during preprocessing (Notebook 3).

```
missing_train = pd.DataFrame({'Missing Count': (adult_data == '?').sum()})
missing_train['Missing %'] = round((missing_train['Missing Count'] / len(adult_data)) * 100,2)
missing_train = missing_train.sort_values(by='Missing Count',ascending=False)
missing_train[missing_train['Missing Count'] > 0]
```

```
missing_test = pd.DataFrame({'Missing Count': (adult_test == '?').sum()})
missing_test['Missing %'] = round((missing_test['Missing Count'] / len(adult_test)) * 100,2)
missing_test = missing_test.sort_values(by='Missing Count',ascending=False)
missing_test[missing_test['Missing Count']>0]
```

### Duplicate Record Findings

A duplicate record check was performed on both datasets.

Results:
- Training Dataset: 24 duplicate records (0.07%)
- Test Dataset: 5 duplicate records (0.03%)

The duplicate rate is very low in both datasets. A decision regarding duplicate removal will be documented in the preprocessing policy notebook.

```
print("Train Duplicates:", adult_data.duplicated().sum())
print("Test Duplicates :", adult_test.duplicated().sum())
```

```
train_dup = round(adult_data.duplicated().sum() / len(adult_data) * 100,2)
test_dup = round(adult_test.duplicated().sum() / len(adult_test) * 100,2)
print("Train Duplicate %:", train_dup)
print("Test Duplicate % :", test_dup)
```
